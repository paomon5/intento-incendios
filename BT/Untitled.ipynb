```python
def compute_scale_and_offset(minValue, maxValue, n):
        """
        Computes the scale_factor and offset for the dataset using a minValue and maxValue, and int n
        """
        # stretch/compress data to the available packed range
        scale_factor = (maxValue - minValue) / (2 ** n - 1)
        # translate the range to be symmetric about zero
        add_offset = minValue + 2 ** (n - 1) * scale_factor
        return scale_factor, add_offset
```

```python
def get_data_pygrib(path_file,name, typeOfLevel=None, level=None):
    """
    Función para extraer los valores de una variable específica contenida en el GFS
    esto funciona sólo para variables 2D el level se pone para cuando se quiera un nivel de presión específico
    """
    if level != None:
        fileidx = pygrib.index(path_file,'name','typeOfLevel', "level")
        grb=fileidx.select(name=name,typeOfLevel = typeOfLevel, level=level)[0]
    else:
        fileidx = pygrib.index(path_file,'name', "typeOfLevel")
        grb=fileidx.select(name=name, typeOfLevel = typeOfLevel)[0] 
        
    data=grb.values
    return data
```
