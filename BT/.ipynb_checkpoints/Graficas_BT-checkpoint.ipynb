```python
# Importación de librerías, se va a usar mucho xarray
from  Modelo_Contiente_RL import Funciones_modelo as FM # esta es la librería creada por nosotros mismos
from netCDF4 import Dataset
import glob
import pandas as pd
import numpy as np
import datetime
import xarray as xr
from cartopy import feature as cfeature
from cartopy import crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import netCDF4 as nc
#import seaborn as sns
import matplotlib as mpl
import cartopy
import cartopy.io.shapereader as shpreader
from skimage import io
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
#from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

import sklearn.metrics as metrics

def genera_mapa_resultado(fecha, tipo = "Big", mapa_salida="binario_orig2D"):
    """Función que genera el mapa 1D o 2D del resultado del modelo de un día en específico
    input:    
    fecha: fecha del día que se quiere analizar (str)
    tipo: Big o Lite 
    mapa_salida =  qué mapa se quiere tener 
    output: 
    pred_y: mapa 1D tal cual lo entrega el modelo (sin la máscara)
    
    mapa_sal: mapa 2D binario tal cual lo entrega el modelo 
    mapa_sal_prob: mapa 2D con los 4 valores de susceptibilidad
    mapa_sal_max: mapa 2D binario con un filtro en la probabilidad de 75%"""
    ruta_output_modelos = "/var/datos_hidrologia/MSI/ContinentalModel/Salidas_Modelo/" + "/{}/{}/"
    total = 201*233
    filenames = [tipo] 
    j = datetime.datetime.strptime(fecha, "%Y-%m-%d")
    for filename in filenames:
        print(fecha)
        #Cargamos Las salidas del modelo guardadas previamente
        if filename =="Lite":
            entradas = np.load(ruta_output_modelos.format(filename, fecha) + "cubo_In_lite_{}.npy".format(fecha))
            mascaras = np.load(ruta_output_modelos.format(filename, fecha) + "cubo_Mask_lite_{}.npy".format(fecha))
            pred_y = np.load(ruta_output_modelos.format(filename, fecha) + "Pred_y_lite_{}.npy".format(fecha))
            pred_y_proba = np.load(ruta_output_modelos.format(filename, fecha) + "Pred_y_proba_lite_{}.npy".format(fecha))
        else:
            entradas = np.load(ruta_output_modelos.format(filename, fecha) + "cubo_In_{}.npy".format(fecha))
            mascaras = np.load(ruta_output_modelos.format(filename, fecha) + "cubo_Mask_{}.npy".format(fecha))
            pred_y = np.load(ruta_output_modelos.format(filename, fecha) + "Pred_y_{}.npy".format(fecha))
            pred_y_proba = np.load(ruta_output_modelos.format(filename, fecha) + "Pred_y_proba_{}.npy".format(fecha))

        resultado_susc=np.zeros_like(pred_y)
        resultado_susc_max=np.zeros_like(pred_y)
        pred_y_total = pred_y
        probabilidades = pred_y_proba[:,1]
        
        ind_falsos = np.where((probabilidades<=0.05))[0]
        ind_baja = np.where((probabilidades>=0.) & (probabilidades<=0.55) & (pred_y_total==1))[0]
        ind_media = np.where((probabilidades>0.55)& (probabilidades<0.75) & (pred_y_total==1))[0]
        ind_alta = np.where((probabilidades>=0.75) & (probabilidades<=1) & (pred_y_total==1))[0]

        #resultado_susc[ind_falsos] = 0
        resultado_susc[ind_baja] = 1
        resultado_susc[ind_media] = 2
        resultado_susc[ind_alta] = 3
        resultado_susc_max[ind_alta] =1
        
        # Acá ya se arma la matriz 2D
        mapa_sal_prob = np.zeros(total)
        mapa_sal_prob[mascaras] = np.nan
        mapa_sal_prob[mapa_sal_prob==0] = resultado_susc.reshape(resultado_susc.shape[0])
        mapa_sal_prob = mapa_sal_prob.reshape(201,233)
        
        mapa_sal = np.zeros(total)
        mapa_sal[mascaras] = np.nan
        mapa_sal[mapa_sal==0] = pred_y.reshape(pred_y.shape[0])
        mapa_sal = mapa_sal.reshape(201,233)
        
        mapa_sal_max = np.zeros(total)
        mapa_sal_max[mascaras] = np.nan
        mapa_sal_max[mapa_sal_max==0] = resultado_susc_max.reshape(pred_y.shape[0])
        mapa_sal_max = mapa_sal_max.reshape(201,233)
        if mapa_salida=="binario_orig2D":
            return mascaras, mapa_sal, pred_y_proba
        if mapa_salida=="binario_orig1D":
            return mascaras, pred_y, pred_y_proba
        if mapa_salida == "binario_orig1D":
            return mascaras, pred_y, pred_y_proba
        if mapa_salida == "susceptibilidad":
            return mascaras, mapa_sal_prob, pred_y_proba
        if mapa_salida == "binario_max2D":
            return mascaras, mapa_sal_max, pred_y_proba
```

```python
def grafica_incendios(fecha, dias=7, tipo="Big", ruta_guardar = None, 
                     lim_lat = [-28,20], lim_lon = [-92, -34], BT= True, DEM=True):
    """Función que grafica la susceptibilidad modelada para un día con la
    antidad de incendios ocurridos en los días siguentes que se le ingresen
    input: 
    fecha: fecha del día de la susceptibilidad (str)
    dias: días hacia adelante que se quieren conocer los incendios
    ocurridos. Si no se ingresa se deja por defecto 7 días que es lo que 
    se supone que predice el modelo.
    tipo: si se quiere la versión 'Big' o 'Lite' del modelo. Por defecto
    está el big"""
    
    #Ruta DEM
    if DEM:
        da = xr.open_rasterio('/var/datos_hidrologia/MSI/ContinentalModel/DEM_SA_05dg/dem_sa_res051.tif')
    #Ruta shape de departamentos
    ruta_dep = "/var/geoespacial/shps/Departamentos/departamentos_WGS84_noSA.shp"
    #Ruta pinciparl del Jupyter
    path = "/home/hidrologia/jupyter/ModeloSusceptibilidadIncendios/AMVA/SIATA_incendios/"
    #Ruta íconos de incendios
    path_ave = path+'reportes/reporte_horario/Tools/'
    path_icono_incendios = path_ave+'icono1.png'

    #Ruta donde se guardan los resultados del modelo
    ruta_output_modelos = "/var/datos_hidrologia/MSI/ContinentalModel/Salidas_Modelo/" + "/{}/{}/"

    #Paleta de colores creada para graficar la topografía
    terrain_mio2 = mpl.colors.ListedColormap(["#01CC66", "#93EE81", "#E9F5BC","#f0e8c3", "#ccbc84", "#bba07c",
                                             "#c3935c", "#b28455", "#806c44", "#997C77", "#B39D99", "#CDBFBC", 
                                             "#E7E0DE", "#F9E9DC", "#FFFFFF"]) ## Colores de la topografía

    ruta_save_root = "/var/datos_hidrologia/MSI/ContinentalModel/Salidas_Modelo/"
    total = 201*233
    filenames = [tipo] ##Modelos entrenados previamente
    j = datetime.datetime.strptime(fecha, "%Y-%m-%d")
    for filename in filenames:
        #fecha = j.strftime("%Y-%m-%d")
        print(fecha)
        cmap_mapas6 = mpl.colors.ListedColormap(["#00000000", '#F7EA00', 
                                                 '#E48900', "#BE0000"])
        #Cargamos Las salidas del modelo guardadas previamente
        mascara, mapa_sal_prob, pred_y_proba= genera_mapa_resultado(fecha, tipo = filename,
                                                                   mapa_salida="susceptibilidad")
        ## Carga de incendios reportados en el tiempo de interés 
        incendios = FM.Carga_incendios_MODIS(j, j + pd.Timedelta(f'{dias}d'))
        incendios = incendios[incendios["probabilidad"]>70]
        incendios.drop(incendios[incendios.latitud <lim_lat[0]].index, inplace=True )
        incendios.drop(incendios[incendios.latitud >lim_lat[1]].index, inplace=True )
        incendios.drop(incendios[incendios.longitud >lim_lon[1]].index, inplace=True )
        incendios.drop(incendios[incendios.longitud <lim_lon[0]].index, inplace=True )
        print("La cantidad de incendios ocurridos durante la siguente semana fueron: ",incendios.shape[0])
        
        
        #Ya se tienen los datos, ya a graficar
        fig=plt.figure(figsize=(19,15))
        gs = gridspec.GridSpec(1, 1)

        ax=fig.add_subplot(gs[0,0],projection=ccrs.PlateCarree())
        ax.set_title("Predicción de incendios para la semana del " + str(fecha), fontsize=18)
        ax.set_extent([lim_lon[1], lim_lon[0], lim_lat[0], lim_lat[1]])
        ax.coastlines("50m",linewidth=1.5)
        ax.add_feature(cartopy.feature.BORDERS)
        ax.add_feature(cartopy.feature.OCEAN)
        #ax.set_title("Pronostico de susceptibilidad a incendios forestales\n para la próxima semana: "+ str(fechas[:10]), 
              #       fontsize=25, color="gray")
            
        #Estos se dejan estáticos porque son las de la matríz 2D
        lat = np.arange(-30,20.25, 0.25)
        lon = np.arange(-92.25,-34, 0.25)
        #niveles = np.arange(0,4,0.01)
        aa = [0,100,200, 300, 400,1000, 2000,3000,4000,5000, 6000,7000]
        norm = mpl.colors.BoundaryNorm(boundaries=aa, ncolors = 15)
        aa2 = [1,2, 3]
        norm2 = mpl.colors.BoundaryNorm(boundaries=aa2, ncolors = 3)
        if DEM:
            trr = ax.contourf(da.x, da.y,da[0], cmap = terrain_mio2, levels = np.arange(0,6000,100), norm =norm,
                  alpha = 0.1, extend = 'max')

        rr = ax.contourf(lon,lat,mapa_sal_prob[::-1], cmap= cmap_mapas6)
        shapefile = list(shpreader.Reader(ruta_dep).geometries())
        ax.add_geometries(shapefile, ccrs.PlateCarree(),edgecolor='#4D4646',facecolor='#FF000000', linewidth=0.5, zorder=4)


#       plt.scatter(incendios["longitud"], incendios["latitud"], color= "#26001B", lw=0.004, marker = "D", alpha =0.6)
        duplas = [list(i) for i in incendios[["longitud", "latitud"]].values]

        icono = (io.imread(path_icono_incendios))
        n_report = 0
        if duplas != None:
            for xx in range(len(duplas)):
                Xcc, Ycc = duplas[xx][0],duplas[xx][1]
                n_report+=1
                im = OffsetImage(icono, zoom=0.25)
                ab = AnnotationBbox(im, (Xcc,Ycc), frameon=False)
                ax.add_artist(ab)
        ## Si se desean graficar las BT de los siguentes días        
        if BT:
            files_bts = sorted(glob.glob(f"/var/meteo/GFS/BT/*{fecha.replace('-','')}*"))
            fechas, plev_values, lat_values, lon_values =read_nc(files_bts[0])
            colores_dias = ["#00207F", "#0033CC", "#0040FF", "#686E7F", "#E2E2EB"] ##Me gusta 
            colores_dias = ["#00207F", "#0033CC", "#0040FF", "#2C61FF", "#686E7F"]## Me gusta

            for i in range(115):
                if i <=23:
                    color = colores_dias[0]
                elif i <=47:
                    color = colores_dias[1]
                elif i <=71:
                    color = colores_dias[2]
                elif i <=95:
                    color = colores_dias[3]
                else:
                    color = colores_dias[4]
                if i%24 == 0:
                    plt.plot(lon_values[i], lat_values[i], color=color,
                         label = pd.to_datetime(fechas[i][0]).strftime("%Y-%m-%d"),)
                else:
                    plt.plot(lon_values[i], lat_values[i], color=color)
                    
            plt.scatter(-75.6,6.5,marker='.',edgecolor='k',s=400,color='yellow', label="Valle de Aburrá", zorder=100)
            #plt.scatter(incendios.longitud, incendios.latitud, c="k", s=0.5, zorder=100, label="Puntos caliente MODIS")
            #gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
            #                  linewidth=1, color='gray', alpha=0.5, linestyle='--')
            plt.legend(fontsize=13, title='Retrotrayectorias para\nlos siguiente días', title_fontsize=15)
            
            
            
            
            
        gl= ax.gridlines(color="gray",draw_labels=True,linestyle="--",lw=1)
        gl.xlabels_top=False
        gl.ylabels_right=False
        gl.xlines=True
        gl.ylines=True
        ax.outline_patch.set_visible(False)

        gl.xlabel_style = {'size': 15} #Tamaño etiquetas eje X 
        gl.ylabel_style = {'size': 15} #Tamaño etiquetas eje Y
        cmap_mapas6 = mpl.colors.ListedColormap(['#F7EA00', '#E48900', "#BE0000"]) # Colores del mapa de susc

        cmap = plt.get_cmap(cmap_mapas6,4)
        cb_ax = fig.add_axes([0.25, 0.02, 0.5, 0.05]) #(X,Y,Largo,Ancho)

        ticks= ['Baja','Media','Alta']
        cbar = mpl.colorbar.ColorbarBase(cb_ax, cmap=cmap, spacing='proportional', orientation='horizontal')
        cbar.set_ticks([0.17,0.5,0.825])
        cbar.set_ticklabels(ticks)
        cbar.set_label("Susceptibilidad a incendios",fontsize=20,color=(0.2, 0.2, 0.4))
        cbar.ax.tick_params(labelsize=20,colors=(0.2, 0.2, 0.4))

        plt.savefig(ruta_guardar+fecha+".png", bbox_inches ='tight')
        
        
def metricas_modelo(fecha, tipo="Big",dias=7, lim_lat = [-28,20], 
                    lim_lon = [-92, -34], mapa_analisis="original"):
    """Función que genera métricas de rendimiento para un día
    input: 
    fecha: fecha del día de la susceptibilidad (str)
    dias: días hacia adelante que se quieren conocer los incendios
    ocurridos. Si no se ingresa se deja por defecto 7 días que es lo que 
    se supone que predice el modelo.
    tipo: si se quiere la versión 'Big' o 'Lite' del modelo. Por defecto
    está el big
    lim_lon y lim_lat: límites de latitud y longitud donde se va a analizar
    el rendimiento del modelo
    
    output: lista con [fecha, recall, accuracy, balanced acuracy, presición, f1 score]"""

    #Ruta donde se guardan los resultados del modelo
    ruta_output_modelos = "/var/datos_hidrologia/MSI/ContinentalModel/Salidas_Modelo/" + "/{}/{}/"

    ruta_save_root = "/var/datos_hidrologia/MSI/ContinentalModel/Salidas_Modelo/"
    total = 201*233
    filenames = [tipo] ##Modelos entrenados previamente
    j = datetime.datetime.strptime(fecha, "%Y-%m-%d")
    for filename in filenames:
        if mapa_analisis=="original":
            mascaras, salida_modelo, pred_y_proba= genera_mapa_resultado(fecha, tipo = filename,
                                                                   mapa_salida="binario_orig2D")
            prediccion = salida_modelo
            
        else:
            #Este entrega un mapa 2D
            mascaras, salida_modelo, pred_y_proba= genera_mapa_resultado(fecha, tipo = filename,
                                                                   mapa_salida="binario_max2D")
            prediccion = salida_modelo      
        
        ## Carga de incendios reportados en el tiempo de interés 
        incendios = FM.Carga_incendios_MODIS(j, j + pd.Timedelta(f'{dias}d'))
        incendios = incendios[incendios["probabilidad"]>70]
        ## Recorte de incendios en el área de interés
        incendios.drop(incendios[incendios.latitud <lim_lat[0]].index, inplace=True )
        incendios.drop(incendios[incendios.latitud >lim_lat[1]].index, inplace=True )
        incendios.drop(incendios[incendios.longitud >lim_lon[1]].index, inplace=True )
        incendios.drop(incendios[incendios.longitud <lim_lon[0]].index, inplace=True )
        real = FM.Obtiene_malla_conteo_incendios_MODIS_XR(incendios, 70, gfs.longitude, gfs.latitude, True)
        real = real.reindex({'lat':real.lat.values[::-1]})
        real = np.array(real)
        ocurridos = real.sum()
        #Esto es un machetazo pero no sé cómo más hacer que los valores diferentes a la máscara se queden 
        real_flat = real.reshape(total)
        real_flat[mascaras] =1000
        real_con_mascara = real_flat[real_flat!=1000]
        
        ## Recorte de a salida del modelo en el área de interés, para poder comparar en 
        ## la misma área que el mapa real que se acabó de crear
        ind_lat = np.where((lat>=lim_lat[1])&(lat<=lim_lat[0]))[0]
        ind_lon = np.where((lon<=lim_lon[0])&(lon>=lim_lon[1]))[0]
        prediccion[ind_lat][:, ind_lon] = np.nan
        #Ahora lo vuelvo pequeño y 1D para ingresarlo a la métricas
        prediccion_flat = prediccion.reshape(total)
        prediccion_flat[mascaras] =1000
        prediccion_flat_mascara = prediccion_flat[prediccion_flat != 1000]  
        prediccion = np.copy(prediccion_flat_mascara)
        #Ahora sí las métricas de desempeño         
        recall = metrics.recall_score(real_con_mascara, prediccion)
        acc = metrics.accuracy_score(real_con_mascara, prediccion)
        bal_acc = metrics.balanced_accuracy_score(real_con_mascara, prediccion)
        prec = metrics.precision_score(real_con_mascara, prediccion, average="binary")
        f1 = metrics.f1_score(real_con_mascara, prediccion)
        
        metricas = [fecha, recall, acc, bal_acc, prec, f1, ocurridos]
    return metricas 
```

```python
def read_nc(file_i):
    """Función que lee los archivos donde se almacenan las BT
    diarias y devuelve cada componente"""
    print(file_i)
    Variable = Dataset(file_i,'r')

    dates  = np.array(Variable.variables['time'][:])

    fechas = pd.to_datetime("1900-01-01 00:00:00") \
                    + pd.to_timedelta(dates, unit='h')

    lon_values = np.array(Variable.variables['lon'][:])
    lat_values = np.array(Variable.variables['lat'][:])
    plev_values = np.array(Variable.variables['level'][:]) #shape 24x241
    fechas = np.array(fechas).reshape(plev_values.shape)
    
    return fechas, plev_values, lat_values, lon_values
```

```python
#Función de pendiente de una recta
def pendiente(x1,y1,x2,y2):
    m = (abs(y2-y1))/(abs(x2-x1))
    return m
```
